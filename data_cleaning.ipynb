```python
import pandas as pd
import numpy as np
import re
```

```python
questions = pd.read_csv('Questions.csv', encoding='ISO-8859-1')
answers = pd.read_csv('Answers.csv', encoding='ISO-8859-1')
tags = pd.read_csv('Tags.csv', encoding='ISO-8859-1')

print(questions.columns, answers.columns, tags.columns)
```

```python
questions
```

```python
questions.drop(['OwnerUserId', 'CreationDate', 'ClosedDate', 'Score'],axis=1, inplace=True)
questions
```

```python
def remove_html_tags(text):
  clean_text= re.sub('<.*?>', '', text)
  return clean_text

questions['Body']= questions['Body'].apply(remove_html_tags)
questions
```

```python
questions.isna().sum()
```

```python
top_10_tags=tags['Tag'].value_counts().head(10).index.tolist()

filtered_tags = tags[tags['Tag'].isin(top_10_tags)]

filtered_tags
```

```python
filtered_tags = filtered_tags.groupby('Id')['Tag'].agg(list).reset_index()
filtered_tags
```

```python
final_data= pd.merge(questions, filtered_tags, on='Id', how='inner')
final_data['Title'] = final_data['Title'].apply(lambda x: x.lower())
final_data['Body'] = final_data['Body'].apply(lambda x: x.lower())
final_data.drop('Id', axis=1, inplace=True)

final_data
```

```python
from google.colab import files

final_data.to_csv('cleaned_data.csv', index=False)

files.download('cleaned_data.csv')
```
